```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Assuming 'data.csv' is your dataset file
df = pd.read_csv('best_buy_laptops_2024.csv')
```

```python
# Display the first few rows of the DataFrame
print(df.head())

```

```python
# Check the shape of the DataFrame (rows, columns)
print(df.shape)
```

```python

# Get basic statistics of the numerical columns
print(df.describe())

```

```python
# Check data types and missing values
print(df.info())
```

```python
# Drop duplicate rows
df = df.drop_duplicates()
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each brand
brand_counts = df['brand'].value_counts()

# Plotting the bar graph
plt.figure(figsize=(10, 6))
brand_counts.plot(kind='bar')
plt.title('Count of Brands')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()

```

```python
# Drop missing values from the 'depth' column
depth_values = df['depth'].dropna()

# Plotting the histogram
plt.figure(figsize=(8, 6))
plt.hist(depth_values, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Depth')
plt.xlabel('Depth')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Drop missing values from both columns
rating_price_data = df[['aggregateRating/ratingValue', 'offers/price']].dropna()

# Sort the data by 'offers/price' for a meaningful line plot
rating_price_data.sort_values(by='offers/price', inplace=True)

# Plotting the line graph
plt.figure(figsize=(10, 6))
plt.plot(rating_price_data['offers/price'], rating_price_data['aggregateRating/ratingValue'], marker='o', linestyle='-')
plt.title('Rating vs Price')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

```python
# Plotting the countplot
plt.figure(figsize=(10, 6))
sns.countplot(x='brand', data=df, palette='viridis')
plt.title('Count of Brands')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Plotting the boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='width', data=df, color='skyblue')
plt.title('Boxplot of Width')
plt.xlabel('Width')
plt.show()
```

```python

# Plotting the boxenplot
plt.figure(figsize=(8, 6))
sns.boxenplot(x='width', data=df, color='skyblue')
plt.title('Boxenplot of Width')
plt.xlabel('Width')
plt.show()
```
