```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
# reading of csv file 
df = pd.read_csv("ASS2_Ice Cream Sales - temperatures.csv")
df.head()
```

```python
# Splitting variables
X = df.iloc[:, :1]  # independent
y = df.iloc[:, 1:]  # dependent
```

```python
x = df['Temperature']
y = df['Ice Cream Profits']
```

```python
# Splitting dataset into test/train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
# y intercept for the regression line
b0=model.intercept_
print(b0)

# coefficient of x for the regression line
b1=model.coef_
print(b1)
```

```python
# Prediction result
y_pred_test = model.predict(X_test)     # predicted value of y_test
y_pred_train = model.predict(X_train)   # predicted value of y_train
```

```python
# Prediction on training set
plt.scatter(X_train, y_train, color = 'orange')
plt.plot(X_train, y_pred_train, color = 'green')
plt.title('Temperature vs Icecream sales (Training Set)')
plt.xlabel('Temperature')
plt.ylabel('Ice-cream sales')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'prediction os sales', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
# Prediction on testing set
plt.scatter(X_test, y_test, color = 'orange')
plt.plot(X_train, y_pred_train, color = 'green')
plt.title('Temperature vs Icecream sales (Testing Set)')
plt.xlabel('Temperature')
plt.ylabel('Ice-cream sales')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'prediction os sales', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,y_pred_test)
```
